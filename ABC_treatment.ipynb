# Сбор таблицы для составления ABC анализа по категории товаров

```python
# Импорт необходимых библиотек

import numpy as np 
import pandas as pd 
```

Загрузим таблицу с категориями: 

```python
category = pd.read_excel('C:/Users/mitrakov/Documents/Справочники/Категории.xlsx') 
category['product'] = category['product'].apply(lambda x: x.strip()) 
category['product'] = category['product'].str.replace(r'^[я]+', '', case=True, regex=True) 
category['product'] = category['product'].apply(lambda x: ' '.join([i.strip().lower() for i in x.split()])) 
category 
```

```python
# Выберем только нужные нам категории товаров: 

category1 = category[(category['category']=='Вет. аптека')|(category['category']=='Груминг')|(category['category']=='Грызуны')|(category['category']=='Кошки')|(category['category']=='Лежанки')|(category['category']=='Миски д/животных')|(category['category']=='Насекомые')|(category['category']=='Переноски')|(category['category']=='Птицы')|(category['category']=='Рептилии')|(category['category']=='Рыбки')|(category['category']=='Собаки')] 
```

```python
# Удалим дубликаты товаров: 

category1 = category1.reset_index(drop=True) 
category1['product1'] = category1[['Code','product']].apply(lambda x: str(x[0])+str(x[1]), axis=1) 
category1 = category1.drop_duplicates(subset=['product1']) 
```

Загрузим таблицу с продажами:

```python
sales = pd.read_excel('C:/Users/mitrakov/Documents/Коммерческий отдел/ABC по 3 позициям розница 2023/Продажа розница.xlsx') 
sales['product'] = sales['product'].apply(lambda x: x.strip()) 
sales['product'] = sales['product'].str.replace(r'^[я]+', '', case=True, regex=True) 
sales['product'] = sales['product'].apply(lambda x: ' '.join([i.strip().lower() for i in x.split()])) 
sales['product1'] = sales[['Code','product']].apply(lambda x: str(x[0])+str(x[1]), axis=1) 
sales 
```

```python
# Сгруппируем по товарам продажи: 

sales = sales.groupby('product1', as_index=False)[['sum_amount','sum_price','sum_sales','profit']].agg({'sum_amount':'sum','sum_price':'sum','sum_sales':'sum','profit':'sum'})
```

Загрузим таблицу с поставщиками: 

```python
provider = pd.read_excel('C:/Users/mitrakov/Documents/Коммерческий отдел/ABC по 3 позициям розница 2023/Поставщики.xlsx') 
provider['product'] = provider['product'].apply(lambda x: str(x).strip()) 
provider['product'] = provider['product'].str.replace(r'^[я]+', '', case=True, regex=True) 
provider['product'] = provider['product'].apply(lambda x: ' '.join([i.strip().lower() for i in x.split()])) 
provider['product1'] = provider[['Code','product']].apply(lambda x: str(x[0])+str(x[1]), axis=1) 
provider 
```

Загрузим таблицу с остатками: 

```python
leftovers = pd.read_excel('C:/Users/mitrakov/Documents/Коммерческий отдел/ABC по 3 позициям розница 2023/Остатки.xlsx') 
leftovers['product'] = leftovers['product'].apply(lambda x: x.strip()) 
leftovers['product'] = leftovers['product'].str.replace(r'^[я]+', '', case=True, regex=True) 
leftovers['product'] = leftovers['product'].apply(lambda x: ' '.join([i.strip().lower() for i in x.split()])) 
leftovers['product1'] = leftovers[['Code','product']].apply(lambda x: str(x[0])+str(x[1]), axis=1) 
leftovers 
```

```python
# Сгруппируем по товарам остатки: 

leftovers = leftovers.groupby('product1', as_index=False)[['amount','sales']].agg({'amount':'sum','sales':'sum'}) 
```

Загрузим таблицу со штрих-кодами: 

```python
barcode = pd.read_excel('C:/Users/mitrakov/Documents/Коммерческий отдел/ABC по 3 позициям розница 2023/Штрих_коды.xlsx') 
barcode['product'] = barcode['product'].apply(lambda x: str(x).strip()) 
barcode['product'] = barcode['product'].str.replace(r'^[я]+', '', case=True, regex=True) 
barcode['product'] = barcode['product'].apply(lambda x: ' '.join([i.strip().lower() for i in x.split()])) 
barcode['product1'] = barcode[['Code','product']].apply(lambda x: str(x[0])+str(x[1]), axis=1) 
barcode 
```

Загрузим таблицу с брендами: 

```python
brand = pd.read_excel('C:/Users/mitrakov/Documents/Коммерческий отдел/ABC по 3 позициям розница 2023/Бренды.xlsx') 
brand['product'] = brand['product'].apply(lambda x: str(x).strip()) 
brand['product'] = brand['product'].str.replace(r'^[я]+', '', case=True, regex=True) 
brand['product'] = brand['product'].apply(lambda x: ' '.join([i.strip().lower() for i in x.split()])) 
brand['product1'] = brand[['Code','product']].apply(lambda x: str(x[0])+str(x[1]), axis=1) 
brand 
```

```python
barcode1 = barcode.groupby('product1', as_index=False)[['product1','barcodes']].agg({'barcodes':list}) 
brand1 = brand.groupby('product1', as_index=False)[['product1','brend','weight_']].agg({'brend':list, 'weight_':'mean'}) 
```

```python
#barcode1['barcode'] = barcode1['barcode'].apply(lambda x: ';'.join(list(set([str(i).replace(' ', '') for i in x if i is not np.nan])))) 

brand1['brend'] = brand1['brend'].apply(lambda x: ';'.join(list(set([str(i).replace(' ', '') for i in x if i is not np.nan])))) 
```

```python
# Товары, которые были в обороте за 12 месяцев

frames = [sales[['product1']], leftovers[['product1']]] 
product = pd.concat(frames).drop_duplicates() 
product
```

```python
product1 = product.merge(category1[['product1','subcategory3','subcategory2','subcategory1','subcategory','category']], how='left', on='product1') 
```

```python
table = product1.merge(sales, how='left', on='product1') 
table.rename(columns={'sum_amount':'Кол-во продаж','sum_price':'Себестоимость продаж','sum_sales':'РТО'}, inplace=True) 
table 
```

```python
table1 = table.merge(barcode1, how='left', on='product1') 
#table1['barcode'] = table1['barcode'].apply(lambda x: ';'.join([str(i) for i in x if i is not np.nan])) 
table1 
```

```python
# Сгруппируем поставщиков по товарам: 

provider1 = provider.groupby('product1', as_index=False)[['product1','purveyor']].agg({'purveyor':set}) 
```

```python
table11 = table1.merge(provider1, how='left', on='product1') 
```

```python
table11 = table11.merge(brand1, how='left', on='product1') 
```

```python
table10 = table11.merge(leftovers, how='left', on='product1').rename(columns={'amount':'Кол-во остатка','sales':'Сумма остатка'}) 
```

Загрузим таблицу со списаниями товаров:

```python
table12 = table10.merge(write_offs, how='left', on='product1').rename(columns={'sum_amount':'Кол-во списания','sum_price':'Сумма списания'}) 
```

Загрузим таблицу с контрактной ценой: 

```python
contract_price = pd.read_excel('C:/Users/mitrakov/Documents/Коммерческий отдел/ABC по 3 позициям розница 2023/Контрактная цена.xlsx') 
contract_price.rename(columns={'price':'Контрактная цена'}, inplace=True) 
contract_price['product'] = contract_price['product'].apply(lambda x: x.strip()) 
contract_price['product'] = contract_price['product'].str.replace(r'^[я]+', '', case=True, regex=True) 
contract_price['product'] = contract_price['product'].apply(lambda x: ' '.join([i.strip().lower() for i in x.split()])) 
contract_price['product1'] = contract_price[['Code','product']].apply(lambda x: str(x[0])+str(x[1]), axis=1) 
contract_price 
```

```python
# Сгруппируем по товарам с ценой: 

p_price = p_price.groupby('product1', as_index=False)[['product1','LastDate','Розничная цена']].agg({'LastDate':'last','Розничная цена':'max'}) 
```

```python
treatment1['РТО 1 месяц на 1 магазин, ШТ'] = treatment1['Кол-во продаж'].divide(treatment1['count_month'] * treatment1['count_stock'], fill_value=0).replace([np.inf, -np.inf], 0).fillna(0) 
```

```python
treatment1['ВД среднемесячный, руб'] = treatment1['profit'] / 12 
```

```python
treatment1['Средняя цена, руб.'] = treatment1['РТО'].divide(treatment1['Кол-во продаж'], fill_value=0).replace([np.inf, -np.inf], 0).fillna(0)  
```

```python
treatment1['Среднесуточная реализация из Супермаг+'] = treatment1['Кол-во продаж'] / 365 
```

```python
treatment1['Пенетрация 1 месяц на 1 магазин'] = treatment1['penetration'].divide(treatment1['count_month'] * treatment1['count_stock'], fill_value=0).replace([np.inf, -np.inf], 0).fillna(0) 
```

```python
treatment1['Коммерческая маржа 1 месяц на 1 магазин, руб'] = treatment1['profit'].divide(treatment1['count_month'] * treatment1['count_stock'], fill_value=0).replace([np.inf, -np.inf], 0).fillna(0) 
```

```python
treatment1['Потери среднемесячные, руб'] = treatment1['Сумма списания'] / 12 
```
